```python
live.drop(['status_id', 'status_published'], axis=1, inplace=True) 
#drop unnecessary columns
```

```python
live.head()
```

```python
X= live
y= live['status_type']
```

```python
#Convert the text values into the numericak (Categories to integers)
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

X['status_type'] = lb.fit_transform(X['status_type'])

y= lb.transform(y)
```

```python
X.info()
```

```python
X.head()
```

```python
columns= X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
ms= MinMaxScaler()
X= ms.fit_transform(X)
```

```python
X= pd.DataFrame(X,columns=[columns])
```

```python
X.head()
```

```python
import warnings 
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans 
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

kmeans.inertia_

```python
labels = kmeans.labels_
correct_labels = sum(y==labels)
print ("RESULTS : %d our of %d samples are correctly labeled." % (correct_labels , y.size))
print ('Accuracy: {0:0.2f}'.format(correct_labels/float(y.size)))
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
correct_labels = sum(y==labels)
print ("RESULTS : %d our of %d samples are correctly labeled." % (correct_labels , y.size))
print ('Accuracy: {0:0.2f}'.format(correct_labels/float(y.size)))
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
correct_labels = sum(y==labels)
print ("RESULTS : %d our of %d samples are correctly labeled." % (correct_labels , y.size))
print ('Accuracy: {0:0.2f}'.format(correct_labels/float(y.size)))
```
